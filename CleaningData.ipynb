```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file = pd.read_csv("/Users/sajeeshsubramanian/Desktop/Sonale/ML_repo/Ex_Files_Machine_Learning_Algorithms/Exercise Files/titanic.csv")
```

```python
file.head()
```

# Filling missing values

```python
file.isnull().sum()
```

```python
#Fill missing age with the mean age of passengers
```

```python
file['Age'].fillna(file['Age'].mean(),inplace = True)
```

```python
file.isnull().sum()
```

```python
file.head(20)
```

```python
#SibSp and Parch could be combined
```

```python
sns.catplot(x='SibSp',y='Survived',data= file,kind = 'point')
```

```python
sns.catplot(x='Parch',y='Survived',data= file,kind = 'point')
```

```python
#Inf :- The more siblings-parent or parent-children there are; the less chances of survival.The fields could be combined
```

```python
file['Comb'] = file['Parch']+file['SibSp']
```

```python
file.head() 
```

```python
#No need to maintain SibSp and Parch
```

```python
file.drop(['SibSp','Parch','PassengerId'],axis=1,inplace=True)
```

```python
file.head()
```

```python
# need to clean cabin column
```

```python
file['Cabin_ind'] = np.where(file['Cabin'].isnull(),0,1)
```

```python
file.head()
```

```python
#convert sex to numeric
```

```python
gender = {'male':0,'female':1}
```

```python
file['Sex']= file['Sex'].map(gender)
```

```python
file.head()
```

```python
#drop unwanted columns
```

```python
file.drop(['Name','Ticket','Cabin','Embarked'],axis=1,inplace=True)
file.head()
```
